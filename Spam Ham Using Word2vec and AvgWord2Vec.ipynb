```python
import pandas as pd
messages=pd.read_csv('SpamClassifier-master/smsspamcollection/SMSSpamCollection',sep='\t',names=["label","message"] )
```

```python
import nltk
import re
nltk.download('stopwords')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
#using Lemmatization
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.split()
    review=[lemmatizer.lemmatize(word.lower(),pos='v') for word in review ]
    review=' '.join(review)
    corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus, messages['message']) if i<1]
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
# training the model from the scratch
model=gensim.models.Word2Vec(words)
```

```python
def avg_word2vec(doc):
    # remove out-of-vocabulary words
    #sent = [word for word in doc if word in model.wv.index_to_key]
    #print(sent)
    
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
                #or [np.zeros(len(model.wv.index_to_key))], axis=0)
```
